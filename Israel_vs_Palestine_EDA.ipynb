### Let's do some transformation and change some string to informative numeric values

```python
def convert_to_numeric(gdp_str):
    if 'million' in gdp_str.lower():
        # Convert millions to numeric
        return float(gdp_str.replace(' million', '')) * 1e6
    elif 'billion' in gdp_str.lower():
        # Convert billions to numeric
        return float(gdp_str.replace(' billion', '')) * 1e9
    else:
        # Handle other cases, or you can choose to return None
        return None
```

```python
# Apply the numeric converter to column
df['GDP (in USD)'] = df['GDP (in USD)'].apply(lambda x: convert_to_numeric(str(x)))
df['Agricultural Output (in USD)'] = df['Agricultural Output (in USD)'].apply(lambda x: convert_to_numeric(str(x)))
df['IT Output (in USD)'] = df['IT Output (in USD)'].apply(lambda x: convert_to_numeric(str(x)))

df.head()
```

```python
df.describe()
```

```python
## change the datatype of some numerical object to float type
for i in ['Population','Active Military Personnel','Reserve Military Personnel','Number of Tanks','Number of Armoured Vehicles']:
    df[i] = df[i].str.replace(',','')
    df[i] = df[i].astype('float')
```

#### Graph which include country to know and compare their details

```python
plt.figure(figsize=(8, 6))

# Get unique countries in the DataFrame
unique_countries = df['Country'].unique()

# Plot each country separately
for country in unique_countries:
    country_data = df[df['Country'] == country]
    plt.plot(country_data['Year'], country_data['GDP Growth Rate (%)'], marker='o', label=country)

plt.title('GDP Growth Rate Over the Years by Country')
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```

### Comparing groth of the country on the bsis of years

```python
years = df['Year']

# Get unique countries in the DataFrame
unique_countries = df['Country'].unique()

# Create subplots for all columns except 'Year' and 'Country' in a single image
columns_to_plot = df.columns[2:]  # Exclude 'Year' and 'Country'
fig, axes = plt.subplots(nrows=len(columns_to_plot), ncols=1, figsize=(8, 6 * len(columns_to_plot)))

for i, column in enumerate(columns_to_plot):
    ax = axes[i]
    for country in unique_countries:
        country_data = df[(df['Country'] == country)]
        ax.plot(country_data['Year'], country_data[column], marker='o', label=country)
    ax.set_title(f'{column} Over the Years')
    ax.set_xlabel('Year')
    ax.set_ylabel(column)
    ax.legend()

plt.tight_layout()
plt.show()
```
